# DataFrame Description :

The age of abalone is determined by cutting the shell through the cone, staining it, and counting the number of rings through a microscope. Attribute Discription:

Sex - M,F, I(infant)

Length - Longest shell measurement

Diamter - Prependicular to length

Height - With meat in shell

Whole weight - Whole Abalone

Shucked weight - Weight of meat

Viscera weight - gut weight (after bleeding)

Shell weight - After being dried

Rings - +1.5 gives the age in next year

# Regression Model building

```python
x_train,x_test,y_train,y_test = train_test_split(m_m_scaled_x,new_rings,test_size = .31, random_state = 1)
```

```python
lr = LinearRegression()
rf = RandomForestRegressor()
xg = XGBRegressor()
sv = SVR()
r = Ridge()
```

# Training the Model

```python
lr.fit(x_train,y_train)
rf.fit(x_train,y_train)
xg.fit(x_train,y_train)
sv.fit(x_train,y_train)
r.fit(x_train,y_train)
```

All models have been trained.
